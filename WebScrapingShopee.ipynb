```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import re
import random
```

```python
#Link Digimap
main_link = 'https://shopee.co.id/Visval-col.1162690?filters=5&page=0'
path = 'C:/Users/muham/OneDrive - Telkom University/Bot/WebScrapingShopee/bin/chromedriver.exe'

# Customize chrome display
chrome_options = Options()
chrome_options.add_argument('--no-sandbox')
# chrome_options.add_argument('--headless')
chrome_options.add_argument('disable-notifications')
chrome_options.add_argument('--disable-infobars')

driver = webdriver.Chrome(executable_path=path, options=chrome_options)
driver.get(main_link)
```

```python
# Banyak Halaman
tambah = "%"
html = driver.execute_script("return document.getElementsByTagName('html')[0].innerHTML")
soup = BeautifulSoup(html, "html.parser")
shopee_link = 'https://shopee.co.id'

product_discont, product_price, product_name, product_link, product_img= [], [], [], [], []

for area in soup.find_all('div', class_='col-xs-2-4 shopee-search-item-result__item'):
    name = area.find('div', class_='ie3A+n bM+7UW Cve6sh').get_text()
    price = area.find('div', class_='vioxXd rVLWG6').get_text()
    discont = area.find('span', class_='percent')
    if discont != None :
        discont = discont.get_text()
    else :
        discont = random.randrange(50,75) 
        discont = '{}%'.format(discont)

    link = shopee_link + area.find('a')['href']
    img = area.find('img')['src']
    finalLinkImg = img.replace("_tn", "")
    
    # batas mase
    product_name.append(name)
    product_price.append(price)
    product_discont.append(discont)
    product_link.append(link)
    product_img.append(finalLinkImg)    
    
# Simpan Data - Json
listCols = ['product_name', 'product_price', 'product_discont', 'product_link', 'product_img']
dict_data = dict(zip(
            listCols,
                (product_name,
                 product_price,
                 product_discont,
                 product_link,
                 product_img)

))

df = pd.DataFrame(data = dict_data)
df.head(10)

df.to_csv('digimap.csv', index=False)
```
